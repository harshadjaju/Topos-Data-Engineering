```python
# import libraries
import requests
import csv
import pandas as pd
import re
from datetime import datetime
from bs4 import BeautifulSoup

# provide URL to scrape
url = 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population'

# hit URL
response = requests.get(url)

# parse HTML page
soup = BeautifulSoup(response.text, 'lxml')
print(soup.prettify())
```

```python
table = soup.find('table', {'class':'wikitable sortable'})
print(table)
```

```python
rows = table.findAll('tr')
print(rows)
```

```python
th = ['2018 Rank', 'City Name', 'State Name', '2018 Estimate', '2010 Census', 'Change (%)', '2016 Land Area (sq mi)', 
          '2016 Land Area (sq km)', '2016 Population Density (per sq mi)', '2016 Population Density (per sq km)', 'Location']
df = pd.DataFrame(columns = th)
df.to_csv('TopCities_byWiki_ToposAssignment.csv', index = False, encoding="utf-8-sig")
```

```python
df
```

```python
all_data = []
links = []
for row in rows:
    # print(row)
    columns = row.findAll('td')
    # print(columns)
    td = []
    if len(columns) > 1:
        city_link = columns[1].find('a')
        #print(city_link.get('href'))
        city_url = 'https://en.wikipedia.org' + city_link.get('href')
        links.append(city_url)
    for column in columns:
        clean_data = column.text.strip()
        td.append(clean_data)
    all_data.append(td)
```

```python
links
```

```python
print(len(links))
print(len(all_data))
```

```python
clist = []
for link in links[:]:
    url = link
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'lxml')
    t = soup.findChildren('table')[0]
    rows = t.find_all('tr',class_ ='mergedrow')
    col_list = []
    cc_list = []
    col = []
    for row in rows:
        col = []
        ch=row.find_all('th')
        cc=[c.text.strip() for c in ch]
        for i in cc:
            i = i.replace('•\xa0','')
            i = re.sub(r'\[.*?\]', '', i)
            col.append(i)
        ch=row.find_all('td')
        cc1=[c.text.strip() for c in ch]
        col_list.append(col)
        cc_list.append(cc1)
    col_lower = [[val.lower() for val in l] for l in col_list]
    cc_lower = [[val.lower() for val in l] for l in cc_list]
    data_dict = [dict(zip(*z)) for z in zip(col_lower, cc_lower)]
    final = {}
    for dd in data_dict:
        final.update(dd)
    clist.append(final)
city_df= pd.DataFrame(clist)
```

```python
data = pd.DataFrame(all_data, columns = th)
data.to_csv('TopCities_byWiki_ToposAssignment.csv', index = False, encoding="utf-8-sig")
```

```python
# read the csv to clean data
city_df = city_df[['mayor', 'summer (dst)', 'area code', 'area codes', 'area code(s)']]
```

```python
final_data = pd.concat([data,city_df], axis = 1)
```

```python
final_data['area codes'] = final_data['area codes'].fillna(final_data['area code(s)'])
```

```python
final_data['area codes'] = final_data['area codes'].fillna(final_data['area code'])
```

```python
final_data.drop(columns = ['area code', 'area code(s)'], inplace = True)
```

```python
final_data.columns = ['2018 Rank','City Name','State Name','2018 Estimate','2010 Census','Change (%)','2016 Land Area (sq mi)', '2016 Land Area (sq km)', '2016 Population Density (per sq mi)', '2016 Population Density (per sq km)', 'Location', 'Mayor', 'Time Zone (Summer)', 'Area Codes']
```

```python
# remove units, commas and any extra data, which might cause any problem during arithmetic calculations
final_data['2016 Land Area (sq mi)'] = final_data['2016 Land Area (sq mi)'].str.replace('sq', '')
final_data['2016 Land Area (sq mi)'] = final_data['2016 Land Area (sq mi)'].str.replace('mi', '')
final_data['2016 Land Area (sq km)'] = final_data['2016 Land Area (sq km)'].str.replace('sq', '')
final_data['2016 Land Area (sq km)'] = final_data['2016 Land Area (sq km)'].str.replace('km2', '')
final_data['2016 Population Density (per sq mi)'] = final_data['2016 Population Density (per sq mi)'].str.replace('/sq', '')
final_data['2016 Population Density (per sq mi)'] = final_data['2016 Population Density (per sq mi)'].str.replace('mi', '')
final_data['2016 Population Density (per sq km)'] = final_data['2016 Population Density (per sq km)'].str.replace('sq', '')
final_data['2016 Population Density (per sq km)'] = final_data['2016 Population Density (per sq km)'].str.replace('/km2', '')

for column in final_data.columns:
    final_data[column] = final_data[column].str.replace(',', '')
    final_data[column] = final_data[column].str.replace('%', '')
```

```python
final_data = final_data.astype(str)
```

```python
for i in range(0, len(final_data['City Name'])):
    #print(city.split('[')[0])
    city = final_data['City Name'][i]
    mayor = final_data['Mayor'][i]
    final_data['City Name'][i] = city.split('[')[0]
    final_data['Mayor'][i] = mayor.split('(')[0]
```

```python
for i in range(0, len(final_data['Location'])):
    location = final_data['Location'][i]
    final_data['Location'][i] = location.split('/')[1]   
```
